## Reading images

```python
from keras.datasets import mnist
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Load the MNIST dataset
(X_train,train_labels),(X_test,test_labels) = mnist.load_data()
print("The shape of X_train dataset is ",X_train.shape)
```

- There are 60000 images with the pixel of 28x28

```python
## selecting the first sample
x  = X_train[1]
print(f"Pixels: {x.shape}")

## Resizing the image
x = cv2.resize(x,(18,18))
x.shape
```

```python
## Image reading
plt.imshow(x,cmap='gray')
```

```python
print(x)
```

- Note: Pixel values lie between 0 and 255
- As we proceed towards the black shade the pixel values i 0
- Towards white it is closer to 255.

```python
## Reading a color image

## Changing the format to jpg format
cat = cv2.imread('cat.jpg')
cat
```

```python
## Reading the array as an image
plt.imshow(cv2.cvtColor(cat,cv2.COLOR_BGR2RGB))
```

```python
print(f'Image has {cat.shape[0:2]} pixles and {cat.shape[-1]} channels')
```

```python
## red cat
cat_r = cv2.imread('cat.jpg')

##Making all the green and blue channels as 0
cat_r[:,:,1:3] =0
plt.imshow(cat_r)
```

```python
## blue cat
cat_b = cv2.imread('cat.jpg')
## Making all the red and green components 0
cat_b[:,:,0:2] = 0
plt.imshow(cat_b)
```

```python
### Green cat
cat_g = cv2.imread('cat.jpg')
cat_g[:,:,0:3:2] = 0
plt.imshow(cat_g)
```
